## Carga de la base datos

```python
import pandas as pd
import numpy as np
envaj = pd.read_csv("envaj_modulos.csv", encoding = 'unicode_escape', dtype='unicode', sep=',')
envaj.head()
```

## Limpieza datos nulos - vacios

```python
#Mean: 26, 32, 33
for column in envaj:
    if(column == "p26" or column == "p32" or column == "p33"):
        envaj[column] = envaj[column].astype(float)
        envaj[column].fillna(value=envaj[column].mean(), inplace=True)
    else:
        envaj[column].fillna(envaj[column].mode()[0], inplace=True)
        envaj[column] = envaj[column].astype(int)
envaj.head()
```
